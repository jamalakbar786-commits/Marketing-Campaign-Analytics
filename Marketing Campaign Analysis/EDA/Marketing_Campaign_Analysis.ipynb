```python
#Box Plot including All Counties

# Although several spending variables contain high-value outliers, these customers represent premium customers rather than data errors. 
# Therefore outliers were retained.
# In marketing datasets, high spenders are often the most valuable customers rather than erroneous observations.

columns_to_plot = [
    "Age", "Income", "Kidhome", "Teenhome", "Recency", 
    "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", 
    "MntSweetProducts", "MntGoldProds", "NumDealsPurchases", 
    "NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases", 
    "NumWebVisitsMonth"
]

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16, 16))
axes = axes.flatten()  

for i, col in enumerate(columns_to_plot):
    sns.boxplot(data=df, y=col, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Distribution of {col}', fontsize=10)
    axes[i].set_ylabel('')  
    
plt.tight_layout()
plt.show()
```

```python
# Feature Engineering for Spending

spend_cols = [
'MntWines','MntFruits','MntMeatProducts',
'MntFishProducts','MntSweetProducts','MntGoldProds'
]

df['Total_Spends'] = df[spend_cols].sum(axis=1)

df['Total_Spends']
```

```python
# Feature Engineering for Purchase

purchase_cols = [
'NumDealsPurchases',
'NumWebPurchases',
'NumCatalogPurchases',
'NumStorePurchases'
]

df['Total_Purchases'] = df[purchase_cols].sum(axis=1)

df['Total_Purchases']
```

```python
#Feature Engineering for Children

df['Childrens'] = df['Kidhome']+df['Teenhome']

df['Childrens']
```

```python
# Feature Engineering for Campaign Accepted

df['Campaign_Acceptance_Counts'] = (
df['AcceptedCmp1']+
df['AcceptedCmp2']+
df['AcceptedCmp3']+
df['AcceptedCmp4']+
df['AcceptedCmp5']
)

df['Campaign_Acceptance_Counts']
```

```python
def segment_High_Income(Income):
    if Income > 75000:
        return 1
    else:
        return 0

df['High_Income'] = df['Income'].apply(segment_High_Income)

df['High_Income']
```

```python
def segment_Young_Customer(Age):
    if Age < 30:
        return 1
    else:
        return 0

df['Young_Customer'] = df['Age'].apply(segment_Young_Customer)

df['Young_Customer']
```

```python
def segment_Campaign_Responder(Response):
    if Response > 0:
        return 1
    else:
        return 0

df['Campaign_Responder'] = df['Response'].apply(segment_Campaign_Responder)

df['Campaign_Responder']
```

```python
def segment_High_Web_Engagement(NumWebVisitsMonth):
    if NumWebVisitsMonth > 5:
        return 1
    else:
        return 0

df['High_Web_Engagement'] = df['NumWebVisitsMonth'].apply(segment_High_Web_Engagement)

df['High_Web_Engagement']
```

```python
def segment_Family_Customer(Childrens):
    if Childrens > 0:
        return 1
    else:
        return 0

df['Family_Customer'] = df['Childrens'].apply(segment_Family_Customer)

df['Family_Customer']
```

```python
p90 = df['Total_Spends'].quantile(0.90)
    
def segment_High_Spender(Total_Spends):
    if Total_Spends > p90:
        return 1
    else:
        return 0

df['High_Spender'] = df['Total_Spends'].apply(segment_High_Spender)

df['High_Spender']
```
